```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to change the path if needed.)
df1 = "Resources/cleaned_airbnb_dataset2.csv"
df2 = "Resources/MeanZIP-3.xlsx - nation.csv"

# Read the School Data and Student Data and store into a Pandas DataFrame
airbnb_df = pd.read_csv(df1)
zipcode_df = pd.read_csv(df2)

# Check names.
airbnb_df.head(10)
```

```python
zipcode_df1 = zipcode_df.rename(columns={"Zip":"zipcode"}, errors="raise")
zipcode_df1
```

```python
# Combine the data into a single dataset
data_complete_df = pd.merge(airbnb_df, zipcode_df1, how="left", on=["zipcode", "zipcode"])
data_complete_df
```

```python
data_complete_df.shape
```

```python
data_complete_df.to_csv("Resources/cleaned_airbnb_dataset3.csv", index=False)
```
