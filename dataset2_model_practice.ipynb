Random Forest Practice

```python
# defining feature
X = airbnb2_df.copy()
X = X.drop("total", axis=1)
X.head()
```

```python
# defining target
y = airbnb2_df["total"].ravel()
y[:5]
```

```python
# splitting into training tests - round 2
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# starting standard scale
scaler = StandardScaler()

# fit with training data
X_scaler = scaler.fit(X_train)

# scale!
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

# research one_hot_encoder and maybe change the label encoder
```

```python
#create the model
rf_model = RandomForestClassifier(n_estimators=128, random_state=78) 
rf_model = rf_model.fit(X_train_scaled, y_train)
prediction = rf_model.predict(X_test_scaled)
# then continue to confusion matrix
```
